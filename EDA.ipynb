```python
#Reading the csv files into dataframes
import pandas as pd
import numpy as np
CabData = pd.read_csv("Cab_Data.csv")
City = pd.read_csv("City.csv")
Customers = pd.read_csv("Customer_ID.csv")
Transactions = pd.read_csv("Transaction_ID.csv")
```

```python
#merging the cab data 
Transactions_Per_city = pd.merge(CabData, Transactions, how='inner')
Transactions_Per_city.dropna(inplace = True)

Transactions_Per_city
```

```python
Transactions_Per_city.to_csv('Transactions_Per_city.csv', sep = ',', index=True)
```

```python
#process of classifying the transactions by the seasons of the year based on the months of each date of travel
# the seasosn will be classified into
#1 Spring(March to May)
#2 Summer(June to August)
#3 Autumn(September to November)
#4 Winter(December to February)
Seasonality=Transactions_Per_city[['Transaction ID','Date of Travel']]
Seasonality
```

```python
Seasonality.rename(columns = {'Date of Travel':'Date_of_Travel'}, inplace = True)
Seasonality
```

```python
Spring=Seasonality[(Seasonality.Date_of_Travel.dt.month > 2) & (Seasonality.Date_of_Travel.dt.month < 6)]
Spring
```

```python
Summer=Seasonality[(Seasonality.Date_of_Travel.dt.month > 5) & (Seasonality.Date_of_Travel.dt.month < 9)]
Summer
```

```python
Autumn=Seasonality[(Seasonality.Date_of_Travel.dt.month > 8) & (Seasonality.Date_of_Travel.dt.month < 12)]
Autumn
```

```python
Winter=Seasonality[(Seasonality.Date_of_Travel.dt.month == 12) | (Seasonality.Date_of_Travel.dt.month ==1)| (Seasonality.Date_of_Travel.dt.month ==2)]
Winter
```

```python
#profits of each company over the time period
#1. First create two dataframes of each of the Cab companies
#2. Find the the difereence between the cost of travel and the price charged
#3. Classify them by City

Income= Transactions_Per_city[['Price Charged','Cost of Trip']]
Income
#Overall profit for each transaction
#Rename the columns to more suitable naming standards
Income.rename(columns = {'Price Charged':'Price_Charged', 'Cost of Trip':'Cost_of_Trip'}, inplace = True)
Profit = Income["Price_Charged"]-Income["Cost_of_Trip"]
Profit

#Total overall profit for both cabs
Total = Profit.sum()
print (Total)
```

```python
#Classifying profits by cab company
Pink_cab_transactions = Transactions_Per_city.loc[Transactions_Per_city['Company'] == 'Pink Cab', ['Price Charged', 'Cost of Trip']]
Pink_cab_transactions
#Rename the columns to more suitable naming standards
Pink_cab_transactions.rename(columns = {'Price Charged':'Price_Charged', 'Cost of Trip':'Cost_of_Trip'}, inplace = True)
#overall profit for the pink cab
Pink_cab_profit = Pink_cab_transactions["Price_Charged"]-Pink_cab_transactions["Cost_of_Trip"]
Pink_cab_profit = Pink_cab_profit.sum()
print(Pink_cab_profit)
```

```python
#Classifying profits by cab company
Yellow_cab_transactions = Transactions_Per_city.loc[Transactions_Per_city['Company'] >= 'Yellow Cab', ['Price Charged', 'Cost of Trip']]
Yellow_cab_transactions
#Rename the columns to more suitable naming standards
Yellow_cab_transactions.rename(columns = {'Price Charged':'Price_Charged', 'Cost of Trip':'Cost_of_Trip'}, inplace = True)
#overall profit for the pink cab
Yellow_cab_profit = Yellow_cab_transactions["Price_Charged"]-Yellow_cab_transactions["Cost_of_Trip"]
Yellow_cab_profit = Yellow_cab_profit.sum()
print(Yellow_cab_profit)
```
